```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
import numpy as np

df=pd.read_excel(r'C:\Users\rodrigo.villanueva.c\Documents\TODO_PYTHON\Curso_Ploty-dash\Datasets\3.6\Info_pais.xlsx')

df.head()
```

```python
dic_color={"North America":"green", "Asia":"yellow", "Europe":"red", "South America":"blue", "Oceania":"brown", "Africa":"orange"}

df["color_cont"]=df["Continente"].map(dic_color)

df.head(100)
```

```python
data=[go.Scatter(
    x=df["Renta per capita"],
    y=df["Esperanza de vida"],
    mode="markers",
    text=df["País"], #cuando pases por arriba aparace el texto pais
    marker=dict(
        #se crea un diccionario con los siguientes valores
        size=(df["Poblacion"])/5000000,
        color=df["color_cont"]
    ))]


layout=go.Layout(title="Renta percapita y Esperanza de Vida",
xaxis=dict(title="Renta per Capita"),
yaxis=dict(title="Esperanza de Vida"),
)


fig=go.Figure(data=data,layout=layout)

pyo.plot(fig, filename="Burbujas_esperanza_vida.html")

```
